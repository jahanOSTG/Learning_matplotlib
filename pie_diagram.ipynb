### Creating Pie Charts

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([45,23,67,34,38,66,43,32])
plt.pie(x)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([40,23,67,34,38,66,43])
name=np.array(["January","February","March","April","May","June","July"])
plt.pie(x,labels=name)
plt.show()
```

### Start Angle
- The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([40,23,67,34,38,66,43])
name=np.array(["January","February","March","April","May","June","July"])
plt.pie(x,labels=name,startangle=90)
plt.show()
```

### Explode

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([40,23,67,34,38,66,43])
name=np.array(["January","February","March","April","May","June","July"])
label=np.array([0.1,0,0,0,0.3,0,0])
plt.pie(x,labels=name,explode=label)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([40,23,67,34,38,66,43])
name=np.array(["January","February","March","April","May","June","July"])
label=np.array([0.1,0,0,0,0.3,0,0.4])
plt.pie(x,labels=name,explode=label,shadow=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([40,23,67,34,38,66,43])
name=np.array(["January","February","March","April","May","June","July"])
label=np.array([0.1,0,0,0,0.3,0,0.4])
clr=np.array(['k','g','cyan','yellow','m','y','k'])
plt.pie(x,labels=name,explode=label,shadow=True,colors=clr)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([40,23,67,34,38,66,43])
name=np.array(["January","February","March","April","May","June","July"])
label=np.array([0.1,0,0,0,0.3,0,0.4])
clr=np.array(['k','g','cyan','yellow','m','y','k'])
plt.pie(x,labels=name,explode=label,shadow=True,colors=clr)
plt.legend(title="Year Rate")
plt.show()
```
