- The colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([7,2,17,2,9,4,11,12,9,6])
y = np.array([88,111,86,103,87,94,78,77,85,86])
colors = np.array([ 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

### Size
- Can change the size of the dots with the s argument.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6,9])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86,55])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75,345])

plt.scatter(x, y, s=sizes)

plt.show()
```

### Alpha

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11])
y = np.array([99,86,87,88,111,86,103,87,94,78])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```
