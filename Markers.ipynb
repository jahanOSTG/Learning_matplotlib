### Matplotlib Marker

```python
import numpy as np
import matplotlib.pyplot as plt
y_point=np.array([5,3,7,8,12,23])
plt.plot(y_point,marker='o')
plt.show()
```

```python
plt.plot(y_point,marker='*')
plt.show()
```

```python
plt.plot(y_point,marker='p')
plt.show()
```

### Marker size

```python
plt.plot(y_point,marker='h',ms=10)
plt.show()
```

### Marker Color
- the keyword argument "markeredgecolor" or the shorter "mec" to set the color of the edge of the markers:

```python
plt.plot(y_point,marker='h',ms=10,mec='r')
plt.show()
```

- the keyword argument "markerfacecolor" or the shorter "mfc" to set the color inside the edge of the markers:

```python
plt.plot(y_point,marker='h',ms=10,mfc='r')
plt.show()
```

```python
plt.plot(y_point,marker='h',ms=10,mfc='r',mec='r')
plt.show()
```
