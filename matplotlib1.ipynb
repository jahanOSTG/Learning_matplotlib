```python
import numpy as np
import matplotlib.pyplot as plt
x_axis=np.array([50,100])
y_axis=np.array([20,120])
plt.plot(x_axis,y_axis)
plt.show()
```

### Plotting Without Line

```python
plt.plot(x_axis,y_axis,'o')
plt.show()
```

```python
plt.plot(x_axis,y_axis,'d')
plt.show()
```

### Multiple Points

```python
import numpy as np
import matplotlib.pyplot as plt
x_axis = np.array([1, 2, 6, 8])
y_axis = np.array([3, 8, 1, 10])
plt.plot(x_axis,y_axis)
plt.show()
```
