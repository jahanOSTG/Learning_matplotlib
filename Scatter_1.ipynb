### Matplotlib Scatter

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([10,5,7,3,10,9,8,3,6,1])
y=np.array([123,56,78,34,100,110,45,23,18,90])
plt.scatter(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([10,5,7,3,10,9,8,3,6,1])
y=np.array([123,56,78,34,100,110,45,23,18,90])
plt.scatter(x,y,color='hotpink')
plt.show()
```

### compare plot

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([5,7,8,7,2,17,2,9,4])
y = np.array([99,86,87,88,111,86,103,87,94])
plt.scatter(x, y)

x = np.array([2,2,8,1,15,8,12,9,7,3,11])
y = np.array([100,105,84,105,90,99,90,95,94,100,79])
plt.scatter(x, y)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([5,7,8,7,2,17,2,9,4])
y = np.array([99,86,87,88,111,86,103,87,94])
plt.scatter(x, y,color='hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11])
y = np.array([100,105,84,105,90,99,90,95,94,100,79])
plt.scatter(x, y,color='cyan')

plt.show()
```

### cannot use the color argument for this, only the c argument.

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([10,5,7,3,10,9,8,3,6,1,5,7,10])
y=np.array([123,56,78,34,100,110,45,23,18,90,77,33,11])
color= np.array(["teal", "lime", "navy", "gold", "salmon", "maroon", "coral", "indigo", "ivory", "chocolate", "slategray", "turquoise", "orchid"
])
plt.scatter(x,y,c=color)
plt.show()
```
